```python
import math as m
import matplotlib.pyplot as plt


x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 1, 2, 3, 4, 5, 6]
z = [m.sin(m.sqrt(x[i] ** 2 + y[i] ** 2)) for i in range(len(x))]

# Создаём 3д разметку
ax = plt.axes(projection="3d")
# Строим график функции на этой разметке
ax.plot3D(x, y, z, 'red')
plt.show()
```
